```python
#examples of break
i=1
while i<=5:
    print(i)
    if(i == 3):
        break  # here the break for stopping
    i+=1
print ("end of this code")
nums = (1, 4, 9, 16, 25, 36, 49, 68, 81, 100 , 81)
x=81
i = 0 #1
while i < len(nums):
    if(nums[i] == x):
        print("number found at :", i)
        break                                     # in this case there is 2 simi terms when loop finds one it stoo thier imedit
    else:
        print("still finding")      
    i += 1 #3
print("end")
```

```python
#examples of continue
i = 0                    #1
while i <= 5:            #2
    if( i == 3):         #3
        i += 1
        continue         # here it remove 3 and print all other   basically it skip
    print(i)             #4
    i += 1               #5

    
```

```python
# if we onle have to print odd numbers
i = 0                    #1
while i <= 10:            #2
    if( i%2 == 0):         #3
        i += 1
        continue         # here it remove 3 and print all other   basically it skip
    print(i)             #4
    i += 1               #5
print (" odd no")


#  for even no 
i = 0                    #1
while i <= 10:            #2
    if( i%2 != 0):         #3
        i += 1
        continue         # here it remove 3 and print all other   basically it skip
    print(i)             #4
    i += 1               #5
print("even no")
```

```python
 # for loops 
# it works when their is sequential type of data like list , tuple, str basically where idx is present

list = [1, 2, 3, 4, 5]

for value in list:
    print(value)

# on tuple
veggies = ("cucumber", "tomato", "ladyfinger", "capsicum")
for valu in veggies:
    print(valu)
print("vegetables")

# for str
str = "apnacollege"
for char in str:
    print(char)
else:
    print("char of this word")
```

```python
# Range function range()

# it return a sequence of numbers, starting from 0 by default , and increments by 1 (by default) , and stop before a specified number

# three types of range 
# start which is optiinal 
# stop compulsory
# step its also opt

print(range(5))
```

```python
#we try to print idx
seq = range(5)
print(seq[0])
print(seq[1])
print(seq[2])
print(seq[3])
print(seq[4])

# if we try to run loop on this sequence 
print("range run thriugh loop")
seq1 = range(5)   # last num is not included like output will be ... 4
for el in seq1:
    print(el)


# another short method to work of range in loop is 
print("easier way to work with loop")
for el in range(10):
    print(el)

```

```python
# three ways of range()
print("for stop")
for el in range(10):                # in this we know are end and start is by default o
    print(el)

print("for start and stop")           # start include stop jot include
for el in range(2,5):
    print(el)

print("for start, stop and step") 
for el in range(2,7,2):               #this range mean start , step stop
    print(el)
```

```python
# to print even by applying range 
print("even numbers")
for el in range(0,20,2):
    print(el)

print("odd numbers")
for el in range(1, 20, 2):
    print(el)
```

```python
 # q2
# Write a program to find the factorial of first n numbers. using for loop

# n = 5
# fact = 1
# i = 1
# while i <= n:
#     fact *= i
#     i+=1
# print("fact of no is:",fact)

# by using for loop

n = 5
fact = 1

for i in range (1, n+1):
    fact *= i

print ("fact value is :",fact)
```
