```python
#slicing
# accessing part of string
# breaking string into parts
# str[starting_idx:ending_idx]    ending index is not included
#agr hm last indx skip kr dain tb bi python smj jay ga last tk jna

str="Women University"
print(str[1:8])  #len(str)  agr hma last len tk jna ho to
print(str[1:]) 
print(str[:8]) 
```

```python
#slicing in negative index
str="Apple"
print(str[-5:-1]) #last one will be neglected like in this ex -1 will ignore
```

```python
light="pink"
if (light=="red"):
    print("stop")
elif(light=="green"):
    print("go")
elif(light=="yellow"):
    print("wait")
else:
    print("light is broken")   # else work at last only and it work when all if and slif statement are false

# we can directly use else after if nut is is important
age = 22
if(age>=18):
    print("can  vote")
else:
    print("cannot vote")

# another example of grading

marks=80
if(marks >= 90):
    grade = "A"
elif(marks >=80  and marks < 90):
    grade = "B"
elif(marks <=70 and marks < 80):
    grade = "C"
else:
    grade = "D"
```
