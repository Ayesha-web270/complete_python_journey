```python
# if we want to overwrite in start we use r+
# read + overwrite(pointer is in start)   no trancate

f = open("sample.txt", "r+") # r+ overwrite in start 

f.write("welcom")

print(f.read()) #file main last pointer jhain ho vahin sa reading start hoti ha 

f.close()
```

```python
# a+ read and append pointer at end no trancate

f = open("sample.txt", "a+") # a+ append  where pointer last

print(f.read()) #file main last pointer jhain ho vahin sa reading start hoti ha 

f.write("abc")

f.close()
```
