```python
# recursive fun 

# 1.what to do   & what are base case would be 

# when function call itself repeadely

def show(n):
    if(n == 0):
        return
    print(n)
    show(n-1)

show(5)
```

```python
def fact(n):
    if (n == 0 or n == 1):
        return 1
    return fact(n-1)* n

print(fact(4))
```

```python
# practice question
# q1
# Write a recursive func to calc sum of first n natura num.


def cal_sum(n):
    if (n == 0):
        return 0
    return cal_sum(n-1) + n

sum = cal_sum(10)
print(sum)
```

```python
# Write a recursive func to print all elements in a list
# |Hint: use list & index as a parameter

def print_list(list,idx=0):
    if (idx == len(list)):
        return
    print(list[idx])
    print_list(list,idx+1)

fruits =["mango","apple","banana"]
print_list(fruits)
```
