```python
# TUPLES IN PYTHON
# A built-in data type that let us creat immutable sequence of values
# we cant change anything frome tuples once we write down like strings
#tuples are created like list but the diffence is [] instead of()

tup=(43,54,13)
print(type (tup))
print(tup[1]) #in tuple we can access indx as well

tup[0]=5 #not allowed

```

```python
# we creat multi type of tuples
tup=()
print(tup) # empty tuple is valid
print(type(tup))

tup1=(1,) # we can use single integer by adding , to it otherwise python thiks it is a simple integer
print(tup1) 
print(type(tup1))

tup1=(1) # we can use single integer by adding , to it otherwise python thiks it is a simple integer and gives that integer type
print(tup1) 
print(type(tup1))

tup1=(1,3,4.5) # its optional for multiple list of tuple to add , on end its completely fine python gives type tuple
print(tup1) 
print(type(tup1))
```

```python
# wap to check if a list contins a palindrom(same and last same as "ma'am" , racecar) of elements.(hint: user copy () method)
# [1,2,3,2,1]     [1."abc","abc",1]
# how to know the palindromic list when we copynit and reverse it  it gives again same list
list1=[1,2,1,2,1]
list2=[1,2,3]

copy_list1=list1.copy()
copy_list1.reverse()

copy_list2=list2.copy()
copy_list2.reverse()


if (copy_list1 == list1):
    print("its palindrom")
if(copy_list2==list2):
    print ("its palindrom")
else:
    print("no its not")
```
