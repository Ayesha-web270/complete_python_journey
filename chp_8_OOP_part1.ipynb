```python
class Student:
    def __init__(self, name, marks1, marks2, marks3):
        self.name = name
        self.marks1 = marks1
        self.marks2 = marks2
        self.marks3 = marks3

    def average(self):
        avg = (self.marks1 + self.marks2 + self.marks3) / 3
        print(f"{self.name} ka average marks hai: {avg}")


s1 = Student("karan", 90, 85, 95)
s1.average()

s2 = Student("mahi", 70, 60, 80)
s2.average()
```

```python
# by the help of list 
class Student:
    def __init__(self, name, marks):   # marks ek list hogi [90, 85, 95]
        self.name = name
        self.marks = marks

    def average(self):
        avg = sum(self.marks) / len(self.marks)
        print(f"{self.name} ka average marks hai: {avg}")


s1 = Student("karan", [90, 85, 95])
s1.average()

# value of attribute is alse changeable

s1.name = "ironman"
s1.average()
```

```python
# static method
# when we dont need of any object 

class Student:
    def __init__(self, name):
        self.name = name

    @staticmethod
    def college_info():     # Object banaye bina call kar sakte ho!
        print("hi! i am learning python")

Student.college_info()   # Object banaye bina call kar sakte ho!
```

```python
# example

class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def multiply(a, b):
        return a * b

print(Calculator.add(5, 3))         # 8
print(Calculator.multiply(4, 6))    # 24
```

```python
# comparison exampla

class Student:
    college_name = "ABC College"
    
    def __init__(self, name):
        self.name = name

    def welcome(self):                    # Instance method
        print(f"Welcome {self.name}")     # object ki value use ki (self.name)

    @staticmethod
    def college_info():                   # Static method
        print("This college was established in 2020")  # koi object value use nahi ki

s1 = Student("karan")
s1.welcome()              # Object zaroori hai (self.name chahiye)
Student.college_info()    # Object zaroori nahi (kuch bhi self se use nahi ho raha)
```

```python
# abstraction

from abc import ABC, abstractmethod

class Vehicle(ABC):        # Abstract class
    @abstractmethod
    def start_engine(self):
        pass                # Sirf naam diya, kaam chhupa diya

class Car(Vehicle):
    def start_engine(self):
        print("Car ka engine chal gaya: Vrooom!")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike ka engine chal gaya: Brrrr!")


c = Car()
c.start_engine()

b = Bike()
b.start_engine()
```

```python
# encapsulation
class Account:
    def __init__(self, balance):
        self.__balance = balance    # __ (double underscore) = private variable

    def show_balance(self):
        print(f"Aapka balance hai: {self.__balance}")

    def deposit(self, amount):
        self.__balance += amount
        print("Paisa deposit ho gaya!")


acc = Account(1000)
acc.show_balance()      # Balance dikhega method ke through
acc.deposit(500)
acc.show_balance()

# Ab direct access karke dekho:
print(acc.__balance)    # ERROR! Direct access nahi kar sakte
```

```python
# examples

"""Create an Account class that has:

2 attributes: balance and account_no
3 methods:

debit (withdraw money)
credit (deposit money)
a method to print the balance """

class Account:
    def __init__(self, account_no, balance):
        # Constructor - jab bhi naya object banega, ye khud chalega
        self.account_no = account_no    # account number ko object ke andar store kiya
        self.balance = balance          # starting balance ko object ke andar store kiya

    def credit(self, amount):
        # Paisa deposit karne wala method
        self.balance += amount          # current balance mein amount add kar diya
        print(f"{amount} credited. New balance: {self.balance}")

    def debit(self, amount):
        # Paisa withdraw karne wala method
        if amount > self.balance:
            # Agar nikalne wali amount, balance se zyada hai to error
            print("Insufficient balance, cannot debit!")
        else:
            self.balance -= amount      # balance mein se amount minus kar diya
            print(f"{amount} debited. New balance: {self.balance}")

    def show_balance(self):
        # Sirf current account details print karta hai
        print(f"Account No: {self.account_no}, Balance: {self.balance}")


# ---- Object create karna ----
acc1 = Account("12345", 1000)   # naya account object bana, account_no="12345", balance=1000

acc1.show_balance()   # abhi ka balance dikhao (1000)
acc1.credit(500)      # 500 deposit karo
acc1.debit(200)       # 200 withdraw karo
acc1.show_balance()   # updated balance dikhao
```
