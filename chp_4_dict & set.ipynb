```python
# 3Dictionary in PYTHON
# store data values in key:values pairs
# unordered (bec in tuple ,list, string their is index but in dict its not),mutable,and dont allow duplicatio keys 
# key should be unique 

info = {
    "key" : "value",
    "name" : "ayesha",
    "learning" : "from apna college",
    "age":22,
    "is_adult":True,
    "marks":90.8,
}
print(info)   # in short we can add multiple datatype in dictionary
```

```python
# we add list and type as well in dict
info = {
   
    "name" : "ayesha",
    "learning" : "from apna college",
    "subjects":["python","html","css"],  # list
    "topics":("dictionary","and ","sets"),   # tuples
    "age":22,
    "is_adult":True,
    "marks":90.8,
}
print(info)   # we can give anything to except tuple because  its imutable normally keys were str
print(type(info)) # Also check type of dict

print(info["subjects"])  # if we access anything which is not availabe it shows error

info["name"]="rafia"#we can change the name as well
info["surname"]="ejaz"#we can assign new key as well
print(info)

# we can add null dictionary as well
null_dict={ }
null_dict["name"]="apna college"
print(null_dict)
```

```python
# Nested dictionary
student={
    "name":"mahi",
    "subjects":{
        "math": 87 ,
        "english":99,
        "science":69,
    
    }
}
print(student)

print(student["subjects"]) #we can access the subject dictionary itself

print(student["subjects"]["math"]) # we cn access multiple keys as wll
```

```python
#Set in PYTHON
# collection of unordered item(no index)
# each element in set must be unique and immutable

# list and dict is not workable in set because they are immutable
collection={1,2,3,4,5,"hello",89.90,2} #set ignres duplicate  its output is unorder thats why no commes 1st in output
print(collection)
print(type(collection))
print(len(collection))

# for creating an empty set in sets 
collection1={}
print(type(collection1)) #it shows dict type 

#thats the correct syntax for creating an empty set in sets
collection3=set()
print(type(collection3))
```
